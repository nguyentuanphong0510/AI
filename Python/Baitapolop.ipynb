```python
import random

# a.
def taomang(m, n):
    return [[random.randint(0, 99) for _ in range(n)] for _ in range(m)]
def xuatmang(a):
    for row in a:
        print(*row)

# b
def xuatdongk(a, k):
    print("Dòng", k, ":", *a[k])

# c
def xuatcotk(a, k):
    print("Cột", k, ":")
    for row in a:
        print(row[k])

# d
def dongtonglonnhat45(a):
    max_sum = -1
    vt = -1

    for i, row in enumerate(a):
        s = sum(row)
        if s > 45 and s > max_sum:
            max_sum = s
            vt = i

    if vt != -1:
        print("Dòng", vt, "có tổng lớn nhất >45:", max_sum)
    else:
        print("Không có dòng nào thỏa")

# e
def cottichnhonhat(a):
    m = len(a)
    n = len(a[0])

    min_tich = None
    vt = -1

    for j in range(n):
        tich = 1
        for i in range(m):
            tich *= a[i][j]

        if min_tich is None or tich < min_tich:
            min_tich = tich
            vt = j

    print("Cột", vt, "có tích nhỏ nhất:", min_tich)

# f
def dongchancotle(a):
    m = len(a)
    n = len(a[0])

    print("Phần tử dòng chẵn, cột lẻ:")
    for i in range(0, m, 2):
        for j in range(1, n, 2):
            print(a[i][j], end=" ")
    print()


#main
m = int(input("Nhập số dòng m: "))
n = int(input("Nhập số cột n: "))

a = taomang(m, n)

print("\nMảng a:")
xuatmang(a)

k = int(input("\nNhập k: "))

# b
xuatdongk(a, k)

# c
xuatcotk(a, k)

# d
dongtonglonnhat45(a)

# e
cottichnhonhat(a)

# f
dongchancotle(a)
```

```python
#Bài 15 Một sinh viên gồm các thông tin sau: Mã sinh viên là chuỗi có 10 ký tự, Tên là chuỗi tối đa 20 ký tự, Năm sinh là một số nguyên ,Điểm trung bình là một số thực a.
print("Bài 15")
#a. Xây dựng cấu trúc SINHVIEN mô tả một sinh viên và Cho một mảng có n sinh viên.
print("a)")
class SinhVien:
  dic_sinhvien = { "masinhvien":[],
                  "tensinhvien":[],
                 "Namsinh":[],
                 "diemtrungbinh":[]
                 }
  def __init__(self,n):
    for i in range(n):
      ma=input("Nhập mã sinh viên ")
      ten=input("Nhập tên sinh viên ")
      nam=int(input("Nhập năm sinh "))
      dtb=float(input("Nhập điểm trung bình "))
      self.dic_sinhvien['masinhvien'].append(ma)
      self.dic_sinhvien['tensinhvien'].append(ten)
      self.dic_sinhvien['Namsinh'].append(nam)
      self.dic_sinhvien['diemtrungbinh'].append(dtb)

SinhVien(2)
print(SinhVien.dic_sinhvien)
#c) Viết hàm cho biết có bao sinh viên đủ điều kiện lên lớp, biết rằng sinh viên đủ điều kiện lên lớp khi điểm trung bình lớn hơn hoặc bằng 5.
print("\nc) ")
print("Danh sách sinh viên đủ điều kiện lên lớp là ")
for i in range(len(SinhVien.dic_sinhvien['diemtrungbinh'])):
  if SinhVien.dic_sinhvien['diemtrungbinh'][i]>=5:
    print(SinhVien.dic_sinhvien['tensinhvien'][i])
#d) Xuất các sinh viên đủ 20 tuổi
print("\nd) ")
print("Danh sách sinh viên 20 tuổi là ")
namhientai = 2026

for i in range(len(SinhVien.dic_sinhvien['Namsinh'])):
    if namhientai - SinhVien.dic_sinhvien['Namsinh'][i] == 20:
        print(SinhVien.dic_sinhvien['tensinhvien'][i])
#e	Đếm số sinh viên học hệ đại học, biết rằng sinh viên hệ DH có mã sinh viên chứa 2 ký tự DH ở vị trí 2,3 trong chuỗi. VD: 02DH0001.
print("\ne) ")
dem = 0

for ma in SinhVien.dic_sinhvien['masinhvien']:
    if len(ma) >= 4 and ma[2:4] == "DH":
        dem += 1

print("Số SV hệ DH:", dem)
#f 	Cho biết trong mảng có bao nhiêu sinh viên có tên «Lan»
print("\nf) ")
dem = 0

for ten in SinhVien.dic_sinhvien['tensinhvien']:
    if "lan" in ten.lower():
        dem += 1

print("Số SV tên Lan:", dem)

#g 	Cho biết trong mảng có bao nhiêu sinh viên có họ «Phan»
print("\ng) ")
dem = 0

for ten in SinhVien.dic_sinhvien['tensinhvien']:
    ho = ten.strip().split()[0].lower()  # lấy họ
    if ho == "phan":
        dem += 1

print("Số SV họ Phan:", dem)
```
