```python
#bài 1

print("Bài 1")
a=[3,9,1,4]
b=[2,7,4,3,2,8]
def tron_mang(a, b):
    c = []
    n = min(len(a), len(b))

    # phần chung
    for i in range(n):
        c.append(min(a[i], b[i]))

    # phần dư
    if len(a) > n:
        c.extend(a[n:])
    else:
        c.extend(b[n:])

    return c
kq = tron_mang(a, b)
print("Mảng kết quả:", kq)
```

```python
#bài2
a = [
    [1, -2, 3],
    [4, 5, -6],
    [7, 8, 9]
]
n = len(a)


#a
def tongtrencheophu(a):
    n = len(a)
    return sum(a[i][j] for i in range(n) for j in range(n) if i + j <= n-1)


#b
def doiam(a):
    for i in range(len(a)):
        for j in range(len(a)):
            if a[i][j] < 0:
                a[i][j] = abs(a[i][j])


#c
def thaychan(a, x):
    for i in range(len(a)):
        for j in range(len(a)):
            if a[i][j] % 2 == 0:
                a[i][j] = x


# d
def toanchan(a):
    return all(x % 2 == 0 for row in a for x in row)


# e
def doixung(a):
    n = len(a)
    return all(a[i][j] == a[j][i] for i in range(n) for j in range(n))


# f
def cheochinhtang(a):
    n = len(a)
    return all(a[i][i] < a[i+1][i+1] for i in range(n-1))


# g
def duoicheophu(a):
    n = len(a)
    for i in range(n):
        for j in range(n):
            if i + j >= n-1:
                print(a[i][j], end=" ")
        print()


#h
def cheophugiam(a):
    n = len(a)
    return all(a[i][n-1-i] > a[i+1][n-2-i] for i in range(n-1))

print("Bài 2")
print("a) Tính tổng các phần tử thuộc tam giác trên của đường chéo phụ kể cả đường chéo phụ trong ma trận vuông a cấp n \n=", tongtrencheophu(a))

doiam(a)
print("\nb) Chuyển các phần tử âm thành trị tuyệt đối của nó trong a \n=", a)

thaychan(a, 0)
print("\nc) Thay các phần tử chẵn trong a bằng số nguyên x cho trước \n=", a)

print("\nd) Kiểm tra a có toàn chẵn không ?\n", toanchan(a))
print("\ne) Cho ma trận vuông a cấp n, viết hàm kiểm tra a có đối xứng không. Biết rằng ma trận đối xứng là ma trận có a[i][j] = a[j][i]\n", doixung(a))
print("\nf) Kiểm tra ma trận vuông a cấp n có đường chéo chính tăng dần không?\n ", cheochinhtang(a))

print("\ng) Xuất các phần tử thuộc tam giác dưới của đường chéo phụ kể cả đường chéo phụ trong ma trận vuông a cấp n")
duoicheophu(a)

print("\nh) Kiểm tra ma trận vuông a cấp n có đường chéo phụ giảm dần không ?\n", cheophugiam(a))
```
