```python
import heapq
import math


def distance(a, b):
    return math.sqrt((a[0] - b[0])**2 + (a[1] - b[1])**2)


def heuristic(current, unvisited, cities):
    if not unvisited:
        return 0
    # lấy khoảng cách nhỏ nhất từ current đến các thành phố chưa thăm
    return min(distance(cities[current], cities[i]) for i in unvisited)


def a_star_tsp(cities):
    n = len(cities)


    pq = []

    start = 0
    heapq.heappush(pq, (0, 0, start, frozenset([start]), [start]))

    while pq:
        f, g, current, visited, path = heapq.heappop(pq)

        # Nếu đã đi hết các thành phố
        if len(visited) == n:
            return path, g

        # Duyệt các thành phố chưa thăm
        for next_city in range(n):
            if next_city not in visited:
                new_visited = visited | {next_city}

                cost = distance(cities[current], cities[next_city])
                new_g = g + cost

                h = heuristic(next_city, set(range(n)) - new_visited, cities)
                new_f = new_g + h

                heapq.heappush(pq, (
                    new_f,
                    new_g,
                    next_city,
                    new_visited,
                    path + [next_city]
                ))

    return None, float('inf')


if __name__ == "__main__":
    # tọa độ các thành phố (x, y)
    cities = [
        (0, 0),
        (1, 5),
        (5, 2),
        (6, 6)
    ]

    path, cost = a_star_tsp(cities)

    print("Đường đi tối ưu:", path)
    print("Chi phí:", cost)
```
